```python
import glob
import numpy as np
```

```python
files = sorted(glob.glob(f'ALL-IDB1/*.jpg'))

f = open("ALL-IDB1_WBC.txt", "w+")

for file in files:
    f.write(file.split("/")[-1].split(".")[-2] + "\n")
    
f.close()
```

```python
watershed_count_file = open("ALL-IDB1_WBC_Count_Watershed.txt", "r")

watershed_count_file_lines = watershed_count_file.readlines()

watershed_count_file_lines = ''.join(watershed_count_file_lines).split("\n")

watershed_counts = []
watershed_counts_f = []

for line in watershed_count_file_lines:
    split = line.split(" ")
    watershed_counts_f += [(split[0], split[1])]
    watershed_counts += [split[1]]
    
    
manual_count_file = open("ALL-IDB1_WBC_Count.txt", "r")

manual_count_file_lines = manual_count_file.readlines()
manual_count_file_lines = ''.join(manual_count_file_lines).split("\n")

manual_counts_f = []
manual_counts = []

for line in manual_count_file_lines:
    split = line.split(" ")
    manual_counts_f += [(split[0], split[1])]
    manual_counts += [split[1]]
    
    
manual_counts = np.asarray(manual_counts).astype(np.int32)
watershed_counts = np.asarray(watershed_counts).astype(np.int32)
```

```python
import math

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


MSE = mean_squared_error(manual_counts, watershed_counts)
MAE = mean_absolute_error(manual_counts, watershed_counts)
R2 = r2_score(manual_counts, watershed_counts)

# print(manual_counts)
# print(watershed_counts)

Accuracy = (1 - (np.abs(np.subtract(manual_counts, watershed_counts)) / manual_counts)) * 100

# l = len(Accuracy)
# i = 0
# while(i < l):
#     if Accuracy[i] == 0:
#         Accuracy = np.delete(Accuracy, i)
#         watershed_counts = np.delete(watershed_counts, i)
#         manual_counts = np.delete(manual_counts, i)
#         len = len - 1
#     i += 1
 
# print(Accuracy.round())

for i in range(len(Accuracy)):
    print(f"{manual_counts_f[i][0]} {manual_counts[i]} {watershed_counts[i]} {Accuracy[i]}")


RMSE = math.sqrt(MSE)


print("Mean Square Error:  ")
print(MSE)
print("Root Mean Square Error: ")
print(RMSE)
print("Mean Absolute Error: ")
print(MAE)
print("R2 Score: ")
print(R2)
print("Accuracy")
print(Accuracy.mean())
```

```python
import numpy as np

def createTexTable(result_fileName, output_fileName, caption = "results", label = "table:result"):
    result_file = open(result_fileName, "r+")
    output_file = open(output_fileName, "w+")
    
    result_lines = result_file.readlines()
    
    columns = result_lines[0][:-1]
    
    columns_header = '| '
    for i in range(len(columns.split('\t'))):
        columns_header += " c |"
    
    columns = " & ".join(columns.split("\t")) + " \\\\"
    
    result_lines = result_lines[1:]
    
    lines = []
    
    for line in result_lines:
        #remove \n at the end of line
        line = line[:-1]
        
        
        
        line_s = line.split('\t')
        
        
        
        #clean the file name
        if '/' in (line_s[0]):
            line_s[0] = line_s[0].split("/")[-1].split(".")[0]
        
        #split with separator
        lines += [line_s]
    
    output = ""
    
    for line in  lines:
        line =  " & ".join(line) + " \\\\" + "\n"
        output += line
        
    
        
    output = "\\begin{table}[H]\n\\centering\n\\begin{tabular}{" + f"{columns_header}" + "}\n\\hline\n" + f"{columns}" + f"\n{output}" + "\\hline\n\\end{tabular}\n\\caption{" + caption + "}\n\\label{" + label + "}\n\\end{table}"
    
    output = output.replace(r"_", r"\_")
        
    print(output)
        
    
result_fileName = "RBC_Results_test.txt"
output = "RBC_Benchmark_table.tex"
    
createTexTable(result_fileName, output)
```
